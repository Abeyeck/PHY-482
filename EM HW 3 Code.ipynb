```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
w=1
x=1
b0=1
v=1
t=np.arange(0,100,1)
a=np.linspace(0,10000000,100)
e = -w*x*a -(b0+a*t)*w*v

plt.plot(t,e)
plt.grid()
plt.xlabel('t')
plt.ylabel('a')
```

We see that the behavior of alpha does changein time, and similarly we would see the same graph with v. So the parameters matter.
