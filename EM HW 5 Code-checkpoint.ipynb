```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
plt.figure(figsize = (12, 8))

R = L = 1


w = np.arange(0, 50, .01)
V = R / np.sqrt(R**2 + ((w**2) * (L**2)))

plt.plot(w, V)
plt.grid()
plt.title('Ratio of Vin/Vout')
plt.xlabel('Frequency')
plt.ylabel('Vout/Vin')
```
