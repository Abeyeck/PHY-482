```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np 
```

```python
beta = 2.0
theta = np.linspace (0.0001,90,1000)

def lmbda(theta, beta):
    x = np.sqrt(1-(np.sin(np.radians(theta))/beta**2))/np.cos(np.radians(theta))
    return x 

ER_EI = (1-lmbda(theta,beta)*beta)/(1+lmbda(theta,beta)*beta)
ET_EI = 2/(1+lmbda(theta,beta)*beta)

```

```python
plt.plot(theta,ER_EI, label = 'ER/EI')
plt.plot(theta,ET_EI, label = 'ET/EI')
plt.grid()
plt.legend()
plt.xlabel("Theta")
plt.ylabel('Value')
```

We see that one of the waves has the same form but the other is flipped from the figure 9.16 model, which makes sense for a perpdenicular polarization case. There is no brewsters angle as neither graph manages to reach such a crital angle.
