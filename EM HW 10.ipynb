```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#assum 1 femtosecond ~ 0 
t=0
λ=500*10**-9
sigma=.35
x=np.linspace(-10,10,1000)
k=(2*np.pi)/λ
R = np.exp((-sigma**2*x**2)/2)*np.cos(k*x)
```

```python
plt.plot(R,color='orange')
plt.grid()
plt.xlabel('X')
plt.ylabel('Value')
```

The wave puslse in the x direction but is restricted within the gaussian packet around it at all times.
