```python
import numpy as np 
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
x = np.arange(-5,5,0.5)
y = np.arange(-5,5,0.5)
X,Y=np.meshgrid(x,y)
vx = X/(X**2+Y**2)**(3/2)
vy = Y/(X**2+Y**2)**(3/2)

plt.figure(figsize=(8,8))
plt.quiver(X,Y,vx,vy)
plt.xlim([-5,5])
plt.ylim([-5,5])
```

```python
x=np.arange(-2.5,2.5,.125)
y=np.arange(-2.5,2.5,.125)
X,Y=np.meshgrid(x,y)
vx = (X + 1)/((X+1)**2 + Y**2) - (X - 1)/((X-1)**2 + Y**2)
vy = Y/((X+1)**2 + Y**2) - Y/((X-1)**2 + Y**2)

plt.figure(figsize=(8,8))
plt.quiver(X,Y,vx,vy)
plt.xlim([-2.5,2.5])
plt.ylim([-2.5,2.5])
```

```python
x=np.arange(-2.5,2.5,.125)
y=np.arange(-2.5,2.5,.125)
X,Y=np.meshgrid(x,y)
vx = (X+2)/((X+2)**2+(Y+2)**2)**(3/2) - (X-2)/((X-2)**2+(Y+2)**2)**(3/2) - (X+2)/((X+2)**2+(Y-2)**2)**(3/2) + (X-2)/((X-2)**2+(Y-2)**2)**(3/2)
vy = (Y+2)/((X+2)**2+(Y+2)**2)**(3/2) - (Y+2)/((X-2)**2+(Y+2)**2)**(3/2) - (Y-2)/((X+2)**2+(Y-2)**2)**(3/2) + (Y-2)/((X-2)**2+(Y-2)**2)**(3/2)

plt.figure(figsize=(8,8))
plt.quiver(X,Y,vx,vy)
plt.xlim([-2.75,2.75])
plt.ylim([-2.75,2.75])
```
